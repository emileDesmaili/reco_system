```python
import pandas as pd
from streamlit_assets.components import get_recos_users
```

### Exploring movielens data for recos

```python
reviews = pd.read_csv('data/movielens/ratings_small.csv',usecols=['userId','movieId','rating'])
reviews.columns=['user_id','item_id','rating']
reviews.describe()
```

```python
movies = pd.read_csv('data/movielens/movies_metadata.csv')
movies = movies.rename(columns={"id": "item_id"})
movies['item_id'] = pd.to_numeric(movies['item_id'], errors='coerce').dropna().apply(int)
```

```python
movie_list = ['Scarface','Batman Begins','Memento','Interstellar']
```

```python
movie_ids = movies.loc[movies['title'].isin(movie_list), 'item_id'].unique()
```

```python
df_recos = get_recos_users(movie_ids,reviews, filter_items=True,filter_n=10, default_rating=4.5)
```

```python
df_recos.head(10)
```

```python
df_merged = movies.merge(df_recos, on='item_id').drop_duplicates(subset='item_id').sort_values(by='match', ascending=False).reset_index(drop=True)
df_merged.sort_values(by='match', ascending=False)
df_merged[['title','item_id','overview','match']]
```
